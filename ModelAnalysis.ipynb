```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
result = pd.read_csv('result.csv')
result = result.to_dict('list')
nifty_list = pd.read_csv('NIFTY50_Sectorwise_List.csv')
result['SECTOR'] = nifty_list['INDUSTRY']
result = pd.DataFrame(result)
result.index = result['COMPANY']
result = result.drop(['COMPANY'],axis=1)
result
```

```python
for index, row in result.iterrows():
    if row['LSTM'] > 1 or row['GRU'] > 1 :
        result.drop(index, inplace=True)
```

```python
for sector in result['SECTOR'].unique():
    result[result['SECTOR']==sector].plot(kind='barh',figsize=(15,4),title=sector,xlabel='RMSE')
```

```python
models = ['LSTM','GRU']
sector_model = pd.read_csv('sector_model.csv')
sector_model.index = sector_model['SECTOR']
sector_model= sector_model.drop(['SECTOR'],axis=1)
for model in models:
    for sector in result['SECTOR'].unique():
        data = result[result['SECTOR']==sector][model]
        avg = round(float(sum(data)/len(data)),4)
        sector_model[model][sector] = avg
        text = 'Analysis of '+str(model)+' results of '+str(sector)+' Sector'
        bar = data.nlargest(5).plot(kind='barh',figsize=(15,5),title=text)
        plt.xlabel('RMSE ( Average ='+str(avg)+')')
        plt.show()
sector_model
```

```python
sector_model.plot(kind='bar',figsize=(20,5))
```

```python
sector_model.to_csv('sector_model.csv')
```
